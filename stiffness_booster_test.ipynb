```python
from preparator import Preparator
from solver import Solver 
import pennylane as qml 


dev = qml.device('lightning.qubit', shots = 1000, wires = 10)

sol = Solver(dev, ansatz = 'CIA', gate = 'Z', preprocessing = '')
prep = Preparator() 

matrix = prep.stiffnessPreparator(10, 100)
minen = sol.checkQUBO(matrix)[0]
sol.solve(matrix, sol = minen, depth = 1, stepsize = 2.0, log = True, stiff_booster = False)
```
