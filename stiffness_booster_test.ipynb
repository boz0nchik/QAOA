```python
from preparator import Preparator
from solver import Solver 
import pennylane as qml 
from pennylane import numpy as np

dev = qml.device('lightning.qubit', shots = 1000, wires = 10)

sol = Solver(dev, ansatz = 'MA-PIA', gate = 'X', preprocessing = '')
prep = Preparator() 

matrix = prep.stiffnessPreparator(10, 100)
#matrix = np.random.uniform(size = (10,10))
#matrix = (matrix.T + matrix) / 2
J,h = sol._isingForm(matrix)
```

```python
print(matrix)
```

```python
minen = sol.checkIsing(J,h)[0]
sol.solve(matrix, sol = minen, depth = 3, stepsize = 0.25, log = True, sboosted = True, sboost = 1000)
```

```python
minen = sol.checkIsing(J,h)[0]
sol.solve(matrix, sol = minen, depth = 3, stepsize = 0.25, log = True, sboosted = False, sboost = 1000)
```
