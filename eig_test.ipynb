```python
from preparator import Preparator
import numpy as np 

prep = Preparator() 

ratio_Q = [] 
ratio_J = [] 

for _ in range (100):

    Q = prep.stiffnessPreparator(size = 10, stiffness = 100000)

    J = prep.isingForm(Q)[0]

    eigQ = np.linalg.eigvals(Q)

    eigJ = np.linalg.eigvals(J)

    modQ = np.abs(eigQ) 
    modJ = np.abs(eigJ)
    
    ratio_Q.append(max(modQ) / min(modQ))
    ratio_J.append(max(modJ) / min(modJ))
```

```python
from matplotlib import pyplot as plt 

plt.scatter(ratio_Q, ratio_J)
```
